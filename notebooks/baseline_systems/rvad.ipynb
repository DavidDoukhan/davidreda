```python
import glob
import pandas as pd
from os.path import splitext, basename
from inaSpeechSegmenter.segmenter import _binidx2seglist
```

```python
# Raw RVAD estimates are obtained in matlab
# see https://github.com/zhenghuatan/rVAD/tree/master/rVAD2.0

def parse_rvad(rvadraw):
    with open(rvadraw) as fid:
        segs = _binidx2seglist(fid.read().strip().split('\n'))
    df = pd.DataFrame.from_records(segs, columns=['label', 'start', 'stop'])
    df.label = df.label.map(lambda x: 'speech' if x == '1' else 'nonspeech')
    df.start = df.start / 100.
    df.stop = df.stop / 100.
    return df
```

```python
for f in glob.glob('../../automatic_baselines/rvad_raw/*.txt'):
    bn = splitext(basename(f))[0]
    print(bn)
    df = parse_rvad(f)
    df.to_csv('../../automatic_baselines/rvad_csv/%s.csv' % bn, index=False)
```
