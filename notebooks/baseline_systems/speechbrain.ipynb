```python
import glob
import pandas as pd
from os.path import basename, splitext
from speechbrain.pretrained import VAD
from pyannote.core import Annotation, Segment
```

```python
VAD = VAD.from_hparams(source="speechbrain/vad-crdnn-libriparty", savedir="pretrained_models/vad-crdnn-libriparty")
```

```python
# replace with corpus wav path on your local machine
wavdir = '/tmp/wav/'
```

```python
def refdur(base):
    df = pd.read_csv('../../annotations/vad/%s.csv' % base)
    return df.stop[len(df) -1]
```

```python
for f in glob.glob(wavdir + '/*.wav'):
    print(f)
    base = basename(splitext(f)[0])
    dur = refdur(base)
    boundaries = VAD.get_speech_segments(f)
    annotation = Annotation()
    annotation[Segment(0, dur)] = 'nonspeech'
    
    for start, stop in boundaries:
        # manage micro window differences in duration
        if abs(dur - stop) < 0.1:
            stop = dur
        seg = Segment(start, stop)
        annotation = annotation.extrude(seg)
        annotation[seg] = 'speech'
        
    anotation = annotation.support(collar = 0.1)
    
    lrec = []
    for seg, _, label in annotation.itertracks(yield_label=True):
        lrec.append((float(seg.start), float(seg.end), label))
    df = pd.DataFrame.from_records(lrec, columns=['start', 'stop', 'label'])
    df.to_csv('../../automatic_baselines/speechbrain_vad/%s.csv' % base, index=False)
    print(lrec)
        
#    annotation[Segment(0, refdur(base))] = 'nonspeech'
```
