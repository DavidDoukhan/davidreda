```python
from inaGVAD.vad_metrics import VadEval
import pandas as pd
```

```python
ve = VadEval(collar=0.3)
```

```python
lsystems = ['inaspeechsegmenter', 'liumspkdiarization_csv', 'pyannote_vad', 'rvad_csv', 'silero_vad', 'speechbrain_vad']
ldf = []
for sys in lsystems:
    _, summary = ve.compare_category('./automatic_baselines/%s/' % sys, 'channel_category', './annotations/filesplit/testset.csv')
    ldf.append(summary)
```

```python
ld = []
for sys, df in zip(lsystems, ldf):
    df = df.set_index('category')
    d = {'system': sys}
    for cat in ['generalist_radio', 'music_radio', 'generalist_tv', 'news_tv']:
        for metric in ['accuracy', 'precision', 'recall']:
            d[(cat, metric)] = df[metric][cat] * 100
    ld.append(d)
df = pd.DataFrame.from_dict(ld).set_index('system')
#df = df * 100

d = {'system' : 'MEAN'}
d.update(df.mean())
dmean = pd.DataFrame.from_dict([d]).set_index('system')
dmean

dall = pd.concat([df, dmean])
#.to_latex()
#df
dall
```

```python
print(dall.to_latex(float_format='%.1f').replace('_', ''))
```

```python
lsystems = ['inaspeechsegmenter', 'liumspkdiarization_csv', 'pyannote_vad', 'rvad_csv', 'silero_vad', 'speechbrain_vad']
ldf = []
for sys in lsystems:
    _, summary = ve.compare_csvset('./automatic_baselines/%s/' % sys, './annotations/filesplit/testset.csv')
    ldf.append(summary)
```

```python
ld = []
for sys, df in zip(lsystems, ldf):
    d = {'system': sys}
    for metric in ['accuracy', 'precision', 'recall']:
        d[metric] = df[metric]
    ld.append(d)
```

```python
print(pd.DataFrame.from_dict(ld).to_latex(float_format='%.2f', index=False).replace('_', ''))
```
